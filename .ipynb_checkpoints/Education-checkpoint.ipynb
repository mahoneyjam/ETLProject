```python
enrollment = df['ENROLL']
```

```python
# Demonstrate calculating measures of central tendency
mean_numpy = np.mean(enrollment)
print(f"The mean enrollment is {mean_numpy}")

median_numpy = np.median(enrollment)
print(f"The median enrollment is {median_numpy}")

mode_scipy = sts.mode(enrollment)
print(f"The mode enrollment is {mode_scipy}")
```
